```python
#Pandas is used for machine learning in the form of DataFrames. Pandas allows for importing and exporting tabular data in various formats, such as CSV or JSON files.
import pandas as pd 
```

```python
#syntax - pandas.read_csv() - read function load the CSV data
#assign the variable"dataset" to load the CSV data
dataset=pd.read_csv("Salary_data.csv")
```

```python
#print the collected data from the client
dataset
```

```python
#split the independent variable alias input data
independent=dataset[["YEARS EXPERIENCE"]]
```

```python
#print the inout data
independent
```

```python
#split the dependent variable alias output data
dependent=dataset[["SALARY"]]
```

```python
#print the output data
dependent
```

```python
#syntax for split train and test : sklearn.model_selection.train_test_split(array,test_size=none,radom_state=none)
from sklearn.model_selection import train_test_split
```

```python
train_test_split(independent, dependent, test_size=.30, random_state=0)
```

```python
x_train,x_test,y_train, y_test=train_test_split(independent, dependent, test_size=.30, random_state=0)
```

```python
#first two lines are procedures
# fit function used to insert the substute value
# fit function - calculate the weight & bais
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# SLR formula ==> y=wx+b
# coef(coefficient)_ function gives array of weights estimated by linear regression
weight=regressor.coef_
weight
```

```python
# SLR formula ==> y=wx+b
# intercept(Constant)_ function gives initial value or starting value or minimum value
bais=regressor.intercept_
bais
```

```python
# cross check the value
y_pred=regressor.predict(x_test)
```

```python
#validating parameter - R square = SSR/SST
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
# pickle function(python tool) - save the final model
import pickle
```

```python
#filename created for finalised model
filename="finalized_model_linear.sav"
```

```python
#dump function use to save the file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load function use to load the saved file
# get the input
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])
result
```
